```python

def check_word_in_sent(word, sentence):
    # lowercase
    # words: list of word
    # sentence: string - stocks_NameSymbolStr_dict
    
    # all_words = re.findall(r'\w+', sentence)
    all_words = [t.text.lower() for t in nlp(sentence)]
    if word.lower() in all_words:
#         print('{}, Found!'.format(word))
        return True
    else:
#         print('{}, Not in sentence!'.format(word))
        return False
```

### Text data processing:

The necessary steps include (but aren’t limited to) the following:
1. Tokenizing sentences to break text down into sentences, words, or other units
2. Removing stop words like “if,” “but,” “or,” and so on


```python
'''
    News data processing:
    - filter out:
        - company name and symbol
        - news sources: website name
        - stopwords
        - delimiters
        
'''

del_words = ['$', '|', 'inc.', 'inc', 'corp', 'corporation', 'company', 'co', 
             'stock', 'stocks', 'nyse', 'nasdaq', 'llc', 'ltd.', 'limited', 'ltd',
             'com', 'brands', 'market']

def get_keywords(data):
    
    d_copy = data.copy()
    d_copy['keywords_list'] = ''
    
    keywords_dict = {}
    keywords_list = []

    for i in range(data.shape[0]):
        s = data.loc[i, 'symbol']

        tmp_news = data.loc[i, 'news_wo_source']

        tmp_keywords = [(token.text.lower(), token.lemma_.lower(), token.pos_) for token in nlp(tmp_news) 
                         if not token.is_stop
                         if not token.is_punct
                         if not token.is_space
                         if not check_word_in_sent(token.text, stocks_NameSymbolStr_dict[s])
                         if token.text.lower() not in del_words
                         and any([c.isalpha() for c in token.text]) == True
                       ]
        
        d_copy.loc[i, 'keywords_list'] = tmp_keywords
        
        if s not in keywords_dict:
            keywords_dict[s] = []

        keywords_dict[s].extend(tmp_keywords)

        keywords_list.extend(tmp_keywords)
    
    return d_copy, keywords_dict, keywords_list
```

### Model:
- features: one-hot encoding for keywords list
- response: y/n for indicating the relateness b/w the news and the stock price change

```python
def get_data_for_model(dat_copy, kw_list):
    tmp_data = dat_copy[['dates', 'symbol', 'news', 'label (p/n)', 'keywords_list']].copy()
    print(tmp_data.shape)
    tmp_data = tmp_data.dropna().reset_index(drop=True)
    print(tmp_data.shape)

    all_keywords_list = sorted(list(dict.fromkeys([k[1] for k in kw_list]).keys()))
    print(len(all_keywords_list))
    
    keywords_most_common_100 = sorted([w[0] for w in Counter([k[1] for k in kw_list]).most_common(100)])

    # all nouns and verbs
    all_keywords_noun_verb_list = sorted(list(dict.fromkeys([k[1] for k in kw_list
                                                             if k[2] in ['VERB', 'NOUN']]).keys()))
    print(len(all_keywords_noun_verb_list))
    
    keywords_most_common_noun_verb_100 = sorted([w[0] for w in Counter([k[1] for k in kw_list
                                                                       if k[2] in ['VERB', 'NOUN']]).most_common(100)])

    return tmp_data, all_keywords_list, keywords_most_common_100, \
            all_keywords_noun_verb_list, keywords_most_common_noun_verb_100
```

```python
tmp_data['all_keywords_encoded'] = ''
tmp_data['keywords_100_encoded'] = ''

X_all = []
X_100 = []
X_all_noun_verb = []
X_100_noun_verb = []

for i in range(tmp_data.shape[0]):
    tmp_k = [k[1] for k in tmp_data.loc[i, 'keywords_list']]

    tmp_all_keywords_index = [all_keywords_list.index(k) for k in tmp_k if k in all_keywords_list]
    tmp_all_keywords_encoded = np.zeros(len(all_keywords_list))
    for c in tmp_all_keywords_index:
        tmp_all_keywords_encoded[c] = 1
    
    tmp_keywords_100_encoded = np.zeros(100)
    tmp_keywords_100_index = [keywords_most_common_100.index(k) for k in tmp_k if k in keywords_most_common_100]
    for k in tmp_keywords_100_index:
        tmp_keywords_100_encoded[k] = 1
       
    X_all.append(tmp_all_keywords_encoded)
    X_100.append(tmp_keywords_100_encoded)
    tmp_data.loc[i, 'all_keywords_encoded'] = tmp_all_keywords_encoded
    tmp_data.loc[i, 'keywords_100_encoded'] = tmp_keywords_100_encoded
    
    # noun and verb
    tmp_all_keywords_nv_index = [all_keywords_noun_verb_list.index(k) for k in tmp_k 
                                if k in all_keywords_noun_verb_list]
    tmp_all_keywords_nv_encoded = np.zeros(len(all_keywords_noun_verb_list))
    for c in tmp_all_keywords_nv_index:
        tmp_all_keywords_nv_encoded[c] = 1
    
    tmp_keywords_nv_100_index = [keywords_most_common_noun_verb_100.index(k) for k in tmp_k
                                if k in keywords_most_common_noun_verb_100]
    tmp_keywords_nv_100_encoded = np.zeros(100)
    for c in tmp_keywords_nv_100_index:
        tmp_keywords_nv_100_encoded[c] = 1
        
    X_all_noun_verb.append(tmp_all_keywords_nv_encoded)
    X_100_noun_verb.append(tmp_keywords_nv_100_encoded)
```

#### Logistic Regression

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()
```

```python
def LR_fun(X_train, y_train, X_test, y_test, features):
    LR = LogisticRegression(random_state=123).fit(X_train, y_train)
    LR_pred = LR.predict(X_test)
    LR_errors = [1-int(r) for r in list(LR_pred == y_test)]
    print('MAE:', round(np.mean(LR_errors), 2), 'degrees.')
    print("Accuracy:",metrics.accuracy_score(y_test, LR_pred))
    print("Precision:",metrics.precision_score(y_test, LR_pred, pos_label='y'))
    print("Recall:",metrics.recall_score(y_test, LR_pred, pos_label='y'))

    mat = confusion_matrix(y_test, LR_pred)
    sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=sorted(y_test.unique()), yticklabels=sorted(y_test.unique()))
    plt.xlabel('true label')
    plt.ylabel('predicted label')
    plt.show()
    plt.close()

    coef = pd.DataFrame(LR.coef_).T
    coef.index = features
    
    
    return LR, LR_pred, coef
```

```python
LR_all, LR_pred_all, coef_all = LR_fun(X_train_all, y_train, X_test_all, y_test, all_keywords_list)

LR_all_aug, LR_pred_all_aug, \
coef_all_aug = LR_fun(X_train_all_aug, y_train_aug, X_test_all_aug, y_test_aug, all_keywords_list)
```

### Random Forest

- Variable importance: quantify the usefulness of all the variables in the entire random forest. It represents how much including a particular variable improves the prediction. 

```python
from sklearn.tree import export_graphviz
import pydot
import graphviz
```

```python
# Random Forest model
def RF_fun(X_train, y_train, X_test, y_test, features):
    RF = RandomForestClassifier(random_state=123).fit(X_train, y_train)
    RF_pred = RF.predict(X_test)
    RF_errors = [1-int(r) for r in list(RF_pred == y_test)]
    print('MAE:', round(np.mean(RF_errors), 2), 'degrees.')
    print("Accuracy:",metrics.accuracy_score(y_test, RF_pred))
    print("Precision:",metrics.precision_score(y_test, RF_pred, pos_label='y'))
    print("Recall:",metrics.recall_score(y_test, RF_pred, pos_label='y'))

    mat = confusion_matrix(y_test, RF_pred)
    sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=sorted(y_test.unique()), yticklabels=sorted(y_test.unique()))
    plt.xlabel('true label')
    plt.ylabel('predicted label')
    plt.show()
    plt.close()

    RF_imps = RF.feature_importances_
    
    feature_imp = sorted([(f, round(imp, 4)) for f, imp in zip(features, RF_imps)], 
                         key = lambda x: x[1], reverse=True)
    
    return RF, RF_pred, feature_imp

# Plot variable importance
def plot_var_imp(n_f, feature_imp):
    plt.style.use('fivethirtyeight')

    plt.barh(np.arange(n_f), [x[1] for x in feature_imp[:n_f]][::-1], align = 'center')
    plt.yticks(np.arange(n_f), [x[0] for x in feature_imp[:n_f]][::-1])
    plt.xlabel('Importance')
    plt.ylabel('Variable')
    plt.title('Variable Importances')
    plt.show()
    plt.close()
```

```python
RF_all, RF_pred_all, feature_imp_all = RF_fun(X_train_all, 
                                              y_train, X_test_all, y_test, all_keywords_list)

plot_var_imp(n_f=20, feature_imp=feature_imp_all)
```

### Naive Bayes Model

It assumes that each feature/variable of the same class makes an:
- independent
- equal

contribution to the outcome.
The assumptions made by NB are not generally correct in real-world situations.


$P(y| X) = \frac{P(X|y) P(y)}{P(X)}$, where $y$ is class variable and $X$ is a dependent feature vector with dimension $d$, i.e. $X = (x_1, x_2, \dots, x_d)$.

Assume that all features in $X$ are mutually independent, conditional on the $y$:
- $P(y|x_1, \dots, x_d) = \frac{P(y) \prod_{i=1}^d P(x_i|y)}{P(x_1)P(x_2) \dots P(x_d)} \propto P(y) \prod_{i=1}^d P(x_i|y)$
- Then, $y = \arg\max_y P(y) \prod_{i=1}^d P(x_i |y)$
